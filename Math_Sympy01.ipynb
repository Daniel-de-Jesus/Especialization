#Uso da biblioteca Sympy
##Esta biblioteca tem a mesma função das bibliotecas Numpy e Scipy, porém ela permite que se associe à uma variável x qualquer uma realidade matemática simbólica, x pode ser tratado como uma variável matemática e se utilizar operações para variáveis contínuas, atribuição de funções, gráficos, etc.
 

##Cabeçalho

```python
#lista de bibliotecas para calculos
#nao entendi o que e essa forma de chamar com o asterisco
from sympy import *
import numpy as np
#Nao sei o que eh essa init_printing() tambem
init_printing(use_unicode=True)
```

```python
#Forca bruta

def limite():
  #Definir o simbolo matematico x
  x= symbols('x')
  #inserir a função para a qual se pretende calcular o limite
  f= (x**4-1)/(x**3-1)
  #associando o limite da funcao a variavel limite
  limite = limit(f, x, 1)
  print("O limite da função é: %f." %limite)
#chamando a funcao limite
limite()
```

```python
x= symbols('x')
print(limit(1/x, x, 0, '+'))
print(limit(1/x, x, 0, '-'))
p1= plot(1/x, xlim = [-10, 10], ylim=[-10,10])
```

```python
p2= plot (sin(x), xlim = [-10,10], ylim=[-10,10])
p3= plot (x+sin(x), xlim = [-10,10], ylim = [-10,10])
```

```python
#colocando os dois plots juntos
p1.append(p3[0])
p2.append(p3[0])
p1.show()
p2.show()
```

```python
# plotando graficos parametricos
t= symbols('t')
plot_parametric ((cos(t), sin(t)), (t, -10, 10), aspect_ratio= [1,1])
```
