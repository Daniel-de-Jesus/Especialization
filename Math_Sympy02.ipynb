#Definindo funções
##Mesmo que a biblioteca permita se identificar um símbolo como uma variável matemática a identificação de um texto e uma função conhecida ainda precisa ser feita, como exemplo temos a função seno de x, sin(x), que pode ser interpretada como uma função seno ou como um texto 'em torno' da variável matemática.
###Vamos fazer uso das seguintes funções e suas derivadas:
    seno
    coseno
    exponencial

```python
from sympy import*
#usando o enconding unicode UTF-8
init_printing(use_unicode= True)
```

```python
#Definir o simbolo matemático
x=symbols('x')
#definir as funcoes matematicas
seno= sin(x)
exponencial=  exp(x)
cosseno= cos(x)
```

###Plotanto os gráficos
    Essa notação simbólica é muito interessante para a aplicação à plots uma
    vez que o matplot entende a função matemática como uma regra e a esquematiza

```python
import matplotlib.pyplot as plt
p1= plot(seno, xlim= [-6,6], ylim=[-1.5,1.5])
p2= plot(cosseno, xlim= [-6,6], ylim=[-1.5,1.5])
p3= plot(exponencial, xlim= [-6,20], ylim=[-1.5,1000])
```

```python
p4= plot(diff(seno, x), xlim= [-6,6], ylim=[-1.5,1.5])
p5= plot(diff(cosseno, x), xlim= [-6,6], ylim=[-1.5,1.5])
p6= plot(diff(exponencial, x), xlim= [-6,20], ylim=[-1.5,1000])
```
