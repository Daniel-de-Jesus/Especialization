```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Lets define the params to figures. Jus two different parameters has been setted.

```python
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
```

```python
df= pd.read_csv("/Salary_Data.csv")
df.head()
```

```python
df.columns=["Experience", "Salary"]
df.info()
```

Basic information about data set

```python
print(f'This dataset has {df.shape[0]} samples and {df.shape[1]} features')
```

Statistics

```python
df.describe()
```

Figures: I'm going to describe every step.

```python
#creates the figure object 
# and set the size of it as 20/12 horizontal/vertical portion

fig= plt.figure(figsize=(20,12))

# add subplots, labeling this firt as the 1 of 4
axis_1 = fig.add_subplot(2,2,1)
# setting the title name
axis_1.set_title('Histogram of Experience (Years)')
#setting the x label
axis_1.set_xlabel('Experience(Years)')
#setting the y label
axis_1.set_ylabel('Frequence')
sns.histplot(data= df, x= 'Experience', ax= axis_1)

# add subplots, labeling this firt as the 2 of 4
axis_2 = fig.add_subplot(2,2,2)
axis_2.set_title('Histogram of Annual Salary($)')
axis_2.set_xlabel('Experience(Years)')
axis_2.set_ylabel('Frequence')
sns.histplot(data= df, x= 'Salary', ax= axis_2, color= 'yellow')

# add subplots, labeling this firt as the 3 of 4
axis_3 = fig.add_subplot(2,2,3)
axis_3.set_title('Experience Boxplot')
axis_3.set_ylabel('Experience(Years)')
# orient='v' --->>> Orientation vertical
sns.boxplot(data= df, x= 'Experience', orient='v', ax= axis_3)

# add subplots, labeling this firt as the 4 of 4
axis_4 = fig.add_subplot(2,2,4)
axis_4.set_title('Salary($)- Boxplot')
axis_4.set_ylabel('Salary (Annual)')
sns.boxplot(data= df, x= 'Salary', orient='v', ax= axis_4, color= 'yellow')
```

Scatter plot to evaluate the correlation between two variables

```python
#sns.scatterplot(data=df, x="Experience", y= "Salary")
sns.regplot(data=df, x="Experience", y= "Salary", ci=99.5)
```
