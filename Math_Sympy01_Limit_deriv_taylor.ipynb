##limite de funções

```python
def limite():
  #Definir o simbolo matematico x
  x= symbols('x')
  #inserir a função para a qual se pretende calcular o limite
  f= (x**4-1)/(x**3-1)
  #associando o limite da funcao a variavel limite
  limite = limit(f, x, 1)
  print("O limite da função é: %f." %limite)
#chamando a funcao limite
limite()
```

```python
x= symbols('x')
print(limit(1/x, x, 0, '+'))
print(limit(1/x, x, 0, '-'))
p1= plot(1/x, xlim = [-10, 10], ylim=[-10,10])
```

```python
#colocando os dois plots juntos
p2= plot (sin(x), xlim = [-10,10], ylim=[-10,10])
p3= plot (x+sin(x), xlim = [-10,10], ylim = [-10,10])
p1.append(p3[0])
p2.append(p3[0])
p1.show()
p2.show()
```

##derivada de funções
    definir a função f
    derivada de f é obtida através de diff(f)
###Primeira derivada

```python
x=symbols('x')
f1= x**2
print(f'A primeira derivada da função quadrática é: {diff(f1)}, como já sabemos')
```

###Segunda derivada

```python
#forma recursiva
print(f'A segunda derivada da função quadrática de forma recursiva é:{diff(diff(f1))}')

#atraves do objeto
print(f'A segunda derivada com o objeto dif(função, variável, ordem) é {diff(f1, x, 2)}')
```

###Derivadas superiores
  diff(função, variável, ordem da derivada)
  Ex.:
    x= symbols('x')

    f= 4*x**4+3*x**3+2*x**2+x-20
    
    primeira derivada= diff(f,x,1)
    
    segunda derivada= diff(f,x,2)
    
    terceira derivada= diff(f,x,3)

```python
x= symbols('x')
f= 4*x**4+3*x**3+2*x**2+x-20
prider= diff(f,x,1)
segder= diff(f,x,2)
terder= diff(f,x,3)
print(f'A primeira, segunda e terceira derivadas da função 4*x**4+3*x**3+2*x**2+x-20 são, respectivamente: {prider}, {segder} e {terder}')
```
