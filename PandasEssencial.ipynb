#OBSERVAÇÃO IMPORTANTE

  Uma Serie assim criada a partir de um dataframe é uma referência aos valores contidos no DataFrame, um ponteiro, e não uma cópia. Assim, se alterarmos o valor na tabela orinal, no DataFrame, o resultado da Serie criada também se altera. Isso é importante para se utilizar em processos em que se pretende fazer cálculos ou estudos de valores em uma tabela dinâmica, mas não em processos que precisam do resultado original. Para se manter o valor original da Serie é necessário criar-se uma cópia da coluna ou linha de interesse. A Serie cópia pode ser utilizada como beckup para os dados.

###aplly( )    ------  >>>>>>>>   DF* e Series

    serie do DataFrame:
      axis= 0 ---> aplica a cada linha
      axis= 1 ---> aplica a cada coluna
  
###map( )      ------  >>>>>>>>   Series
    
    Esta função aplica um metodo funcioal a uma serie. O retorno é uma serie cópia.
    serie.map(função)

###applymap( ) ------  >>>>>>>>   DF

    Esta função aplica um metodo funcional à cada termo do DataFrame. O retorno é um DataFrame cópia.
